# Merging
merg the to dataframes based on common key column

### Syntax
- df.merge(df1,df2, on = "Column_Name", how="type of join" )

### Types of join
- inner
- outer
- left
- right 
- cross

```python
import pandas as pd

#customer dataframe
df_customers = pd.DataFrame({
    'CustomerID':[1,2,3],
    'Name':['Ramesh','Suresh','Kalpesh']
})

#order dataframe
df_orders = pd.DataFrame({
    'CustomerID':[1,2,4],
    'OrderAmount':[250,450,350]
})
```

```python
df_merged= pd.merge(df_customers, df_orders, on= "CustomerID", how="inner")
print("Inner Join:")
print(df_merged)
```

```python
df_merged= pd.merge(df_customers, df_orders, on= "CustomerID", how="outer")
print("Outer Join:")
print(df_merged)
```

```python
df_merged= pd.merge(df_customers, df_orders, on= "CustomerID", how="left")
print("Left Join:")
print(df_merged)
```

```python
df_merged= pd.merge(df_customers, df_orders, on= "CustomerID", how="right")
print("Right Join:")
print(df_merged)
```

```python
df_merged= pd.merge(df_customers, df_orders, how="cross")
print("Cross Join:")
print(df_merged)
```

# Concatination
- combine the data in rows or columns 

# Syntax
- pd.concatinate(df1,df1, axis=0,ignore_index= True)

```python
#region1
df_Region1 = pd.DataFrame({
    'CustomerID':[1,2],
    'Name':['Gopal','Raju']
})

#region2
df_Region2 = pd.DataFrame({
    'CustomerID':[3,4],
    'Name':['Shyam','Baburao']
})
```

```python
# concatinate dataframes
df_concat = pd.concat([df_Region1, df_Region2], axis=0, ignore_index=True)
print("Concatenated DataFrames:")
print(df_concat)
```

```python
# concatinate dataframes
df_concat = pd.concat([df_Region1, df_Region2], axis=1, ignore_index=True)
print("Concatenated DataFrames:")
print(df_concat)
```
