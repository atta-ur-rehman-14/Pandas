### How can we create dataframe using dictionary and save them

```python
import pandas as pd
```

### Creating dataframe using Dictinary 

```python
Data = {
    "Name": ["Atta", "Ali", "Ahmad"],
    "Age": [22,24,22],
    "City": ["Karachi", "Lahore", "Chakwal"]
}
```

```python
df = pd.DataFrame(Data)
print(df)
```

### Saving in CSV File
#### to csv method

```python
df.to_csv("output.csv", index=False) 
# index=False is used to prevent the index column from being written to the CSV file.
```

### How to save Excel File

```python
df.to_excel("output.xlsx", index=False)
```

### How to save Json File

```python
df.to_json("output.json", orient="records")
```

### Filtering Rows

```python
filtered_row = df[df["ORDERNUMBER"] > 10100]
print(filtered_row)
```

```python
filter = df[(df["ORDERNUMBER"]>10100) & (df["QUANTITYORDERED"] > 30)]
print(filter)
```

```python
fil = df[(df["ORDERNUMBER"] > 10000) | (df["YEAR_ID"] > 2024)]
print(fil)
```
