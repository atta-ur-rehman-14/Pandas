# sorting
- sorted rows in dataframe based on values of 1 or more column
### Syntax
- df.sort_value(by="Column_Name",True,implace=True) 
- if sort in Ascending order user True
- if sort in descending order user False

## Sorting data in one column

```python
import pandas as pd

data = {
    "Name": ["Atta","kamran","bilal"],
    "Age": [44,30,25],
    "salary": [50000,60000,70000]
}
df= pd.DataFrame(data)
print(df)


```

```python
df.sort_values(by="Age", ascending=True, inplace=True)
print(df)
```

```python
df.sort_values(by="Name", ascending=False, inplace=True)
print(df)
```

```python
df.sort_values(by=["Age","salary"], ascending=[True,False], inplace=True)
print(df)
```

# Aggregation

## Summary statics


```python
avg_salary = df["salary"].mean()
print("Average Salary:", avg_salary)
salary_sum = df["salary"].sum()
print("Total Salary:", salary_sum)
salary_count = df["salary"].count()
print("Count of Salary:", salary_count)
Age_max = df["Age"].max()
print("Maximum Age:", Age_max)
Name_min= df["Name"].min()
print("Minimum Name:", Name_min)
```

# Grouping
- divide data in to categories

```python
import pandas as pd

data = {
    "Name": ["Atta","kamran","bilal","Saim","Jafar"],
    "Age": [25,30,25,35,30],
    "salary": [50000,60000,70000,80000,90000]
}
df= pd.DataFrame(data)
print(df)
```

```python
grouped = df.groupby("Age")["salary"].sum()
print(grouped)
```

```python
grouped = df.groupby("Age")["salary"].mean()
print(grouped)
```

```python
grouped = df.groupby(["Age","Name"])["salary"].sum()
print(grouped)
```
