## Adding Column

```python
import pandas as pd

data = {
    "Name" : ["Atta","Ali ","Ammar","Hassan","Waqas","Haider"],
    "Age" : [22,21,23,34,25,26],
    "salary" : [10000,40000,50000,22000,30000,35000],
    "performance_score" : [50,70,20,44,80,55]
}

df= pd.DataFrame(data)
print(df)

```

```python
df.to_csv("employee_data.csv", index=False)
```

## Adding column via assigning

### Syntax
- df["Column_Name"] = adding_new_data_value

```python
df["Bonus"] = df["salary"] * 0.1  # 10% increment in salary
print(df)
```

## Using Insert Method
- store column in a specific index or position
### Syntax
- df.insert(loc,"Column_Name",data)

```python
df.insert(0,"EmployeeID",[10,20,30,40,50,60])
print(df)
```

# Adding Data

## .loc[] Method
### Syntax
- df.loc[row_index, "Column_Name"] = new_value

```python
df.loc[0, "salary"] = 15000
print(df)
```

### Update multiple values

```python
df["salary"] = df["salary"] * 1.05
print(df)
```

# Removing Data

## Drop Method
### Syntax
- df.drop(columns = ["Column_Name"],inplace= True)

### Single column

```python
df.drop(columns = ["performance_score"], inplace = True)
print(df)
```

### Multiple column

```python
df.drop(columns = ["Age","Bonus"], inplace = True)
print(df)
```

```python
df["Bonus"] = df["salary"] * 0.1
print(df)
```
