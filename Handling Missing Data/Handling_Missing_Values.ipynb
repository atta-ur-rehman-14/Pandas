## isnull() Method
- Nan --> value is missing
- None --> value is present

```python
import pandas as pd

df = pd.read_csv("Messy_Teen_Mental_Health_Dataset.csv")
print(df)
```

```python
print(df.isnull())
```

```python
print(df.isnull().sum())
```

# Handling missing values

### Hoe can we drop missing data

```python
df.dropna(axis=0, inplace=True)
```

```python
df.fillna(0, inplace=True)
print(df)
```

```python
df["sleep_hours"].fillna(df["sleep_hours"].mean(), inplace=True)
print(df)
```

# Interpolation Method
- used to handle missing values and predict the estimate and workwith number
- prserve data integrity
- smooth trends
- avoid data loss
- linear method
- polynomial method
- time method

df.interpolate(method = "linear", axis = 0, inplace = True)
- axis=0 --> 0 for rows 
- axis=1 --> 1 for columns

```python
df["daily_social_media_hours"] = df["daily_social_media_hours"].interpolate(method="linear")
print(df)
```
